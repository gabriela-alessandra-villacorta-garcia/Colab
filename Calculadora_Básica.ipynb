```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def power(a, b):
    return a**b

def divide(a, b):
    return a / b

def calculator():
    while True:
        print("Seleccione la operación:")
        print("1. Suma")
        print("2. Resta")
        print("3. Multiplicación")
        print("4. Potenciación")
        print("5. División")
        print("6. Salir")

        try:
            choice = int(input("Ingrese una opción de operación o salida (1-6): "))
        except ValueError:
            print("Opción inválida. Ingrese un número del 1 al 6.")
            continue

        if choice == 6:
            print("Saliendo de la calculadora.")
            break

        if choice not in (1, 2, 3, 4, 5):
            print("Opción inválida. Por favor, seleccione una opción válida.")
            continue

        num1 = float(input("Ingrese el primer número: "))
        num2 = float(input("Ingrese el segundo número: "))

        if choice == 1:
            print("Resultado:", add(num1, num2))
        elif choice == 2:
            print("Resultado:", subtract(num1, num2))
        elif choice == 3:
            print("Resultado:", multiply(num1, num2))
        elif choice == 4:
            print("Resultado:", power(num1, num2))
        elif choice == 5:
            if num2 != 0:
                print("Resultado:", divide(num1, num2))
            else:
                print("Error: División por cero no permitida.")

calculator()
```
